The universal approximation theorem states that a neural network with one hidden layer can approximate any continuous function. However, there is no guarantee that a 'perfect' neural network achieves 100% accuracy if the function we aim to approximate is non-continuous. The function we aim to learn is $T(s)$, where $s$ is the spin configuration and $T$ is the temperature. Since phase transitions are characterized by non-analytic and thus non-continuous points (e.g the magnetization is linear in $s$, but as a function of $T$ non-continuous at the critical point $T_{c}$), this might also impact the continuity of $T(s)$ such that we never achieve 100% accuracy.
